```python
df2
```

## 3. 将薪水规格化为形如1000-2000格式

```python
df2
```

### 3.1 剔除sal为空的行

```python
df3 = df2.loc[df2['sal'].notnull(),:]
```

### 3.2 规范化sal数据

参考：https://blog.csdn.net/weixin_45221012/article/details/105517276

```python
df3['sal'].unique()
```

```python
# 数据表增加两列['min_salary']和['max_salary']，初值同['Salary']
df3['min_salary'] = df3['sal']
df3['max_salary'] = df3['sal']
```

```python
df3
```

```python
# 使用"字符串包含函数"——str.contains
# 使用正则函数——re.findall(r'[\d\.]+', '待查找字符串')
# 对"万/月"进行处理
import re
df3.loc[df3['sal'].str.contains('万/月'), 'min_salary'] = df3[df3['sal'].str.contains('万/月')]['min_salary'].apply(lambda x:float(re.findall(r'[\d\.]+', x)[0]))
df3.loc[df3['sal'].str.contains('万/月'), 'max_salary'] = df3[df3['sal'].str.contains('万/月')]['max_salary'].apply(lambda x:float(re.findall(r'[\d\.]+', x)[1]))
```

```python
df3
```

```python
# 对"万/年"进行处理
# 年按12薪
# 四舍五入函数round
df3.loc[df3['sal'].str.contains('万/年'), 'min_salary'] = df3[df3['sal'].str.contains('万/年')]['min_salary'].apply(lambda x:round(float(re.findall(r'[\d\.]+', x)[0]) / 12, 1))
df3.loc[df3['sal'].str.contains('万/年'), 'max_salary'] = df3[df3['sal'].str.contains('万/年')]['max_salary'].apply(lambda x:round(float(re.findall(r'[\d\.]+', x)[1]) / 12, 1))
```

```python
df3
```

```python
# 对"千/月"进行处理
df3.loc[df3['sal'].str.contains('千/月'), 'min_salary'] = df3[df3['sal'].str.contains('千/月')]['min_salary'].apply(lambda x:round(float(re.findall(r'[\d\.]+', x)[0]) / 10, 1))
df3.loc[df3['sal'].str.contains('千/月'), 'max_salary'] = df3[df3['sal'].str.contains('千/月')]['max_salary'].apply(lambda x:round(float(re.findall(r'[\d\.]+', x)[1]) / 10, 1))
```

```python
df3
```

```python
# 验证
df3['min_salary'].unique()
```

```python
df3['max_salary'].unique()
```

```python
# 发现还要对 元/天、元/小时 进行处理
```

```python
# 对日薪"元/天"的最低最高薪水，均按22天
df3.loc[df3['sal'].str.contains('元/天'), 'max_salary'] = df3[df3['sal'].str.contains('元/天')]['max_salary'].apply(lambda x:round(float(re.findall(r'[\d\.]+', x)[0]) * 22 / 10000, 1))
df3.loc[df3['sal'].str.contains('元/天'), 'min_salary'] = df3[df3['sal'].str.contains('元/天')]['max_salary']
```

```python
df3
```

```python
# 对时薪"元/小时"的最低最高薪水，均按22天 × 8小时
df3.loc[df3['sal'].str.contains('元/小时'), 'max_salary'] = df3[df3['sal'].str.contains('元/小时')]['max_salary'].apply(lambda x:round(float(re.findall(r'[\d\.]+', x)[0]) * 22 * 8 / 10000, 1))
df3.loc[df3['sal'].str.contains('元/小时'), 'min_salary'] = df3[df3['sal'].str.contains('元/小时')]['max_salary']
```

```python
# 验证
df3['min_salary'].unique()
```

```python
df3['max_salary'].unique()
```

```python
df3
```
